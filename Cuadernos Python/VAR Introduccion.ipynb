## 🔎   Vector autoregresivo VAR(p)
- Un modelo VAR (vector autoregresivo) es generalmente utilizado cuando se desean encontrar relaciones simultaneas en un grupo de variables, permitiendo así estimar un sistema de ecuaciones, a diferencia de los modelos ARIMA tradicionales que permiten estimar sólo una ecuación.

- El modelo VAR es especialmente útil cuando las relaciones entre variables se transmiten a varios periodos, corrigiendo los problemas de endogeneidad (simultaneidad entre variables) y mostrando la interrelación dinámica que hay entre las variables estacionarias.

- El modelo VAR, propuesto por Christopher Sims en 1980, es una extensión multivariada de los modelos autorregresivos univariados. En un VAR, cada variable es una función lineal de sus propios valores rezagados y de los rezagos de todas las demás variables en el sistema.

**Características Clave:**

Simplicidad: Todos los variables son tratadas simétricamente sin necesidad de distinguir entre variables endógenas y exógenas.

Flexibilidad: Capaz de capturar relaciones dinámicas complejas.

Aplicabilidad: Utilizado en economía, finanzas, meteorología y otras disciplinas.


## 🔢   Modelo VAR($p$)

La serie multivariada  $\boldsymbol{z_t}$ sigue un modelo  VAR de orden $p$, VAR($p$), si

$$\boldsymbol{z}_{t}=\phi_{0}+\sum_{i=1}^{p} \phi_{i} z_{t-i}+\boldsymbol{a}_{t}$$

- $\phi_0$:= es una constante de dimensión $k$

- $\phi_i$  son matrices de  $k \times k$  para $i > 0$,  $\phi_p \neq \boldsymbol{0}$

- $\boldsymbol{a_t}$ es una sucesión  de vectores aleatorios independientes e identicamente distribuidos (iid) con media cero y matriz de covarianza $\sum_a$ definida positiva.




### VAR($1$)


$$\boldsymbol{z}_{t}=\boldsymbol{\phi}_{0}+\boldsymbol{\phi}_{1} \boldsymbol{z}_{t-1}+\boldsymbol{a}_{t}$$

Escribiendo el modelo explicitamente se tiene:


$$\begin{pmatrix} z_{1t} \\
z_{2t}
\end{pmatrix}
=
\begin{pmatrix}
\phi_{10} \\
\phi_{20}
\end{pmatrix}
+
\begin{pmatrix}
\phi_{1,11} & \phi_{1,12} \\
\phi_{2,11} & \phi_{2,12}
\end{pmatrix}
\begin{pmatrix}
z_{1,t-1} \\
z_{2,t-1}
\end{pmatrix}
+
\begin{pmatrix}
a_{1t} \\
a_{2t}
\end{pmatrix}$$





$$\begin{array}{l}
z_{1 t}=\phi_{10}+\phi_{1,11} z_{1, t-1}+\phi_{1,12} z_{2, t-1}+a_{1 t} \\
z_{2 t}=\phi_{20}+\phi_{1,21} z_{1, t-1}+\phi_{1,22} z_{2, t-1}+a_{2 t}
\end{array}$$


- Las ecuaciones muestran que la variable $z_{1t}$ es explicada por rezagos de $z_{2t}$ y de sí misma; y de igual manera la variable $z_{2t}$ es afectada por rezagos de $z_{1t}$ y de sí misma, de manera que no es posible distinguir entre variables endógenas y variables exógenas .

- Un VAR($1$) se refiere a un modelo cuyo máximo rezago es $1$.




## 👁️‍🗨️ Notas

- En otras palabras, bajo el marco de Granger, decimos
que $z_{1t}$ causa $z_{2t}$ si la información pasada de $z_{1t}$ mejora el pronóstico de $z_{2t}$

- Para el modelo bivariado VAR(1) en la ecuación,  si $\sum_a$ no es una matriz diagonal, entonces $z_{1t}$ y $z_{2t}$ se correlacionan instantáneamente (o contemporáneamente
correlacionado). 
- En este caso, $z_{1t}$ y $z_{2t}$ tienen causalidad de Granger instantánea. La causalidad instantánea va en ambos sentidos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.diagnostic import het_arch, acorr_ljungbox
from statsmodels.stats.diagnostic import normal_ad
from scipy.stats import jarque_bera
```

- Análisis de relaciones entre PIB (GDP) y tasa de desempleo (UNE).


```python
# Cargar los datos
url = "https://raw.githubusercontent.com/Wilsonsr/Series-de-Tiempo/main/Data/blanchQua.csv"
datos = pd.read_csv(url)
datos['fecha'] = pd.date_range(start='1948-04-01', periods=len(datos), freq='QE')
datos.set_index('fecha', inplace=True)
```

```python
# Crear las series de tiempo
gdp = datos['GDP']
une = datos['U']
```

```python
# Graficar
datos[['GDP', 'U']].plot(title="PIB y Desempleo")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 8))
ax[0].plot(datos['GDP'], label='PIB', color='blue')
ax[0].set_title('PIB')
ax[0].legend()
ax[1].plot(datos['U'], label='Desempleo', color='orange')
ax[1].set_title('Desempleo')
ax[1].legend()
plt.show()
```

Prueba de  Dickey Fuller


\begin{cases}
H_0: z_t \  \text{es  no  estacionaria}\\
H_1: z_t \ \text{es estacionaria}
\end{cases}

rechazar $H_0  \ \text{ si p-value} < \alpha$

```python
# Prueba de raíz u nitaria (ADF)
def adf_test(serie, nombre):
    result = adfuller(serie, autolag='BIC')
    print(f"ADF Test para {nombre}:")
    print(f"  Estadístico: {result[0]:.4f}, p-valor: {result[1]:.4f}")
    print("  ¿Estacionaria?:", "Sí" if result[1] < 0.05 else "No")
    print("-" * 50)

```

```python

adf_test(une, "Desempleo (U)")
adf_test(gdp, "PIB (GDP)")
```

```python
# Crear dataset bivariado
data_bv = datos[['GDP', 'U']].dropna()
```

```python

# Selección del rezago óptimo
modelo_var = VAR(data_bv)
resumen = modelo_var.select_order(maxlags=12)
print("Criterios de información:")
print(resumen.summary())
```

```python
# Estimación del modelo VAR con 2 rezagos
modelo = modelo_var.fit(2)
print(modelo.summary())
```

```python
# Prueba de estabilidad (raíces dentro del círculo unitario)
print("Raíces del modelo VAR:")
print(modelo.roots)
```

```python
abs(modelo.roots)
```

```python
modelo.resid
```

###  Aplica Ljung-Box a cada serie de residuos

```python

from statsmodels.stats.diagnostic import acorr_ljungbox

resid = modelo.resid


for col in resid.columns:
    lb_test = acorr_ljungbox(resid[col], lags=[8], return_df=True)
    print(f"{col}: p-value = {lb_test['lb_pvalue'].values[0]}")

```

### Normalidad

```python
#  Normalidad (Jarque-Bera)
print("\nTest de normalidad (Jarque-Bera):")
for col in modelo.resid.columns:
    jb = jarque_bera(modelo.resid[col])
    print(f"{col}: p-valor = {jb.pvalue:.4f}")
```

```python
print(modelo.is_stable())  # True si cumple el supuesto
```

###  Heterocedasticidad (ARCH test)

```python

print("\nTest ARCH (heterocedasticidad):")
for col in modelo.resid.columns:
    print(f"{col}:")
    arch_result = het_arch(modelo.resid[col])
    print(f"  p-valor: {arch_result[1]:.4f}")
```

###  Causalidad de Granger

```python

print("\nCausalidad de Granger:")
print("GDP causa UNE:")
print(modelo.test_causality(causing='GDP', caused='U', kind='f'))
print("UNE causa GDP:")
print(modelo.test_causality(causing='U', caused='GDP', kind='f'))
```

# Funciones de Impulso-Respuesta (IRF)

📘 ¿Qué es una Función de Impulso-Respuesta (IRF)?

La Función de Impulso-Respuesta (IRF, por sus siglas en inglés: Impulse Response Function) es una herramienta fundamental en el análisis de modelos VAR (Vector Autoregressive), que permite estudiar cómo una variable responde a un shock (perturbación inesperada) en otra variable del sistema a lo largo del tiempo.

🧠 Interpretación intuitiva:
La IRF muestra cómo cambia una variable dependiente tras un shock unitario en una variable explicativa, manteniendo todas las demás condiciones constantes.

⚙️ ¿Qué mide exactamente?
El impacto dinámico de una perturbación en una variable sobre sí misma o sobre otras variables del modelo.

La respuesta se mide durante varios períodos futuros (por ejemplo, días, meses o trimestres).

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

irf = modelo.irf(20)

fig = irf.plot(impulse='U', response='GDP')
fig.set_size_inches(6, 4)  
plt.title("IRF: Impacto de U sobre GDP")
plt.tight_layout()
plt.show()
```

🧠 Lectura de la curva:
- En el periodo 0, el impacto inicial del shock en el desempleo sobre el PIB es negativo, indicando una caída inmediata del PIB.

- En los  periodos (1 a 6), el PIB responde de forma positiva, alcanzando un máximo alrededor del periodo 5, lo que podría interpretarse como una corrección del efecto inicial.

- Después del pico, el efecto comienza a disminuir gradualmente, acercándose a cero, lo que indica que el impacto del shock en U sobre GDP es transitorio.

- Las líneas punteadas representan los intervalos de confianza 95%.

# Descomposición de la varianza

Qué es la Descomposición de la Varianza (FEVD)?

La Descomposición de la Varianza del Error de Pronóstico (Forecast Error Variance Decomposition, FEVD) permite cuantificar qué proporción de la incertidumbre (varianza) en el pronóstico de una variable se debe a los shocks propios y a los shocks de las demás variables del modelo VAR.

📈 ¿Qué muestra?
Por ejemplo, si estamos pronosticando el PIB (GDP), el FEVD indica cuánto de su variabilidad futura se explica por shocks en el PIB mismo, en el desempleo (U), u otras variables del sistema.

El análisis se realiza para distintos horizontes de tiempo (por ejemplo, 1, 5, 10 periodos hacia adelante).

```python
fevd = modelo.fevd(10)
fevd.plot()
plt.show()
```

- En los primeros periodos, casi el 100% de la variabilidad del pronóstico del PIB se explica por shocks en el mismo PIB.

A medida que pasan los periodos, una pequeña proporción (alrededor del 10%) comienza a explicarse por shocks en el desempleo (U).

Esto indica que el PIB es fuertemente autoexplicativo, y los shocks en la tasa de desempleo tienen un efecto limitado pero creciente sobre su incertidumbre futura.

- En el periodo 0, más del 50% de la incertidumbre sobre el desempleo se explica por shocks externos (principalmente del PIB).

- A medida que transcurren los periodos, los shocks propios de GDP (barras negras) ganan relevancia y explican cerca del 70% de la variabilidad futura.

- Esto sugiere que, en el corto plazo (2 periodos), el desempleo es influenciado por el mismo y  otras variables del sistema (como el PIB) 50-50% , pero con el tiempo por shocks propios(30%) y  externos (70%).

```python

# Pronóstico (8 trimestres)
forecast = modelo.forecast_interval(data_bv.values[-2:], steps=8)
forecast_df = pd.DataFrame(forecast[0], columns=['GDP', 'U'])
forecast_df.index = pd.date_range(start=datos.index[-1] + pd.offsets.QuarterEnd(), periods=8, freq='Q')
forecast_df.plot(title="Pronóstico de GDP y U")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Hacer pronóstico con intervalos de confianza
forecast_mean, lower, upper = modelo.forecast_interval(data_bv.values[-modelo.k_ar:], steps=8)

# 2. Usar solo las columnas que entrenaron el modelo
cols = data_bv.columns  # ← típicamente ['GDP', 'U']

# 3. Crear DataFrames para pronóstico e intervalos
forecast_df = pd.DataFrame(forecast_mean, columns=cols)
lower_df = pd.DataFrame(lower, columns=cols)
upper_df = pd.DataFrame(upper, columns=cols)

# 4. Crear índice de fechas para el pronóstico
forecast_index = pd.date_range(start=datos.index[-1] + pd.offsets.QuarterEnd(), periods=8, freq='QE')
forecast_df.index = lower_df.index = upper_df.index = forecast_index

# 5. Ajuste del modelo (valores ajustados del VAR)
fitted_df = pd.DataFrame(modelo.fittedvalues, index=datos.index[modelo.k_ar:], columns=cols)

# 6. Crear gráficos
fig, axes = plt.subplots(2, 1, figsize=(10, 7), sharex=True)

for i, var in enumerate(cols):
    # Serie observada
    axes[i].plot(datos.index, datos[var], label='Observado', color='black')

    # Ajuste del modelo (in-sample)
    axes[i].plot(fitted_df.index, fitted_df[var], label='Ajuste VAR', color='blue', linestyle='dotted')

    # Pronóstico
    axes[i].plot(forecast_df.index, forecast_df[var], label='Pronóstico', color='orange', linestyle='--')

    # Intervalo de confianza
    axes[i].fill_between(forecast_df.index, lower_df[var], upper_df[var],
                         color='orange', alpha=0.2, label='IC 95%')

    axes[i].set_title(f'Variable: {var}')
    axes[i].legend()

# Título general
plt.suptitle("Pronóstico con Ajuste VAR e Intervalos de Confianza", fontsize=14)
plt.tight_layout()
plt.show()
```
