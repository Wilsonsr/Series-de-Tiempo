## üîé   Vector autoregresivo VAR(p)
- Un modelo VAR (vector autoregresivo) es generalmente utilizado cuando se desean encontrar relaciones simultaneas en un grupo de variables, permitiendo as√≠ estimar un sistema de ecuaciones, a diferencia de los modelos ARIMA tradicionales que permiten estimar s√≥lo una ecuaci√≥n.

- El modelo VAR es especialmente √∫til cuando las relaciones entre variables se transmiten a varios periodos, corrigiendo los problemas de endogeneidad (simultaneidad entre variables) y mostrando la interrelaci√≥n din√°mica que hay entre las variables estacionarias.

- El modelo VAR, propuesto por Christopher Sims en 1980, es una extensi√≥n multivariada de los modelos autorregresivos univariados. En un VAR, cada variable es una funci√≥n lineal de sus propios valores rezagados y de los rezagos de todas las dem√°s variables en el sistema.

**Caracter√≠sticas Clave:**

Simplicidad: Todos los variables son tratadas sim√©tricamente sin necesidad de distinguir entre variables end√≥genas y ex√≥genas.

Flexibilidad: Capaz de capturar relaciones din√°micas complejas.

Aplicabilidad: Utilizado en econom√≠a, finanzas, meteorolog√≠a y otras disciplinas.


## üî¢   Modelo VAR($p$)

La serie multivariada  $\boldsymbol{z_t}$ sigue un modelo  VAR de orden $p$, VAR($p$), si

$$\boldsymbol{z}_{t}=\phi_{0}+\sum_{i=1}^{p} \phi_{i} z_{t-i}+\boldsymbol{a}_{t}$$

- $\phi_0$:= es una constante de dimensi√≥n $k$

- $\phi_i$  son matrices de  $k \times k$  para $i > 0$,  $\phi_p \neq \boldsymbol{0}$

- $\boldsymbol{a_t}$ es una sucesi√≥n  de vectores aleatorios independientes e identicamente distribuidos (iid) con media cero y matriz de covarianza $\sum_a$ definida positiva.




### VAR($1$)


$$\boldsymbol{z}_{t}=\boldsymbol{\phi}_{0}+\boldsymbol{\phi}_{1} \boldsymbol{z}_{t-1}+\boldsymbol{a}_{t}$$

Escribiendo el modelo explicitamente se tiene:


$$\begin{pmatrix} z_{1t} \\
z_{2t}
\end{pmatrix}
=
\begin{pmatrix}
\phi_{10} \\
\phi_{20}
\end{pmatrix}
+
\begin{pmatrix}
\phi_{1,11} & \phi_{1,12} \\
\phi_{2,11} & \phi_{2,12}
\end{pmatrix}
\begin{pmatrix}
z_{1,t-1} \\
z_{2,t-1}
\end{pmatrix}
+
\begin{pmatrix}
a_{1t} \\
a_{2t}
\end{pmatrix}$$





$$\begin{array}{l}
z_{1 t}=\phi_{10}+\phi_{1,11} z_{1, t-1}+\phi_{1,12} z_{2, t-1}+a_{1 t} \\
z_{2 t}=\phi_{20}+\phi_{1,21} z_{1, t-1}+\phi_{1,22} z_{2, t-1}+a_{2 t}
\end{array}$$


- Las ecuaciones muestran que la variable $z_{1t}$ es explicada por rezagos de $z_{2t}$ y de s√≠ misma; y de igual manera la variable $z_{2t}$ es afectada por rezagos de $z_{1t}$ y de s√≠ misma, de manera que no es posible distinguir entre variables end√≥genas y variables ex√≥genas .

- Un VAR($1$) se refiere a un modelo cuyo m√°ximo rezago es $1$.




## üëÅÔ∏è‚Äçüó®Ô∏è Notas

- En otras palabras, bajo el marco de Granger, decimos
que $z_{1t}$ causa $z_{2t}$ si la informaci√≥n pasada de $z_{1t}$ mejora el pron√≥stico de $z_{2t}$

- Para el modelo bivariado VAR(1) en la ecuaci√≥n,  si $\sum_a$ no es una matriz diagonal, entonces $z_{1t}$ y $z_{2t}$ se correlacionan instant√°neamente (o contempor√°neamente
correlacionado). 
- En este caso, $z_{1t}$ y $z_{2t}$ tienen causalidad de Granger instant√°nea. La causalidad instant√°nea va en ambos sentidos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.diagnostic import het_arch, acorr_ljungbox
from statsmodels.stats.diagnostic import normal_ad
from scipy.stats import jarque_bera
```

El conjunto de datos se encuentra, por ejemplo, en algunos tutoriales en R o Python para ilustrar:

- Series macroecon√≥micas trimestrales de Estados Unidos.

- An√°lisis de relaciones entre PIB (GDP) y tasa de desempleo (UNE).


```python
# Cargar los datos
url = "https://raw.githubusercontent.com/Wilsonsr/Series-de-Tiempo/main/Data/blanchQua.csv"
datos = pd.read_csv(url)
datos['fecha'] = pd.date_range(start='1948-04-01', periods=len(datos), freq='QE')
datos.set_index('fecha', inplace=True)
```

```python
# Crear las series de tiempo
gdp = datos['GDP']
une = datos['U']
```

```python
# Graficar
datos[['GDP', 'U']].plot(title="PIB y Desempleo")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 8))
ax[0].plot(datos['GDP'], label='PIB', color='blue')
ax[0].set_title('PIB')
ax[0].legend()
ax[1].plot(datos['U'], label='Desempleo', color='orange')
ax[1].set_title('Desempleo')
ax[1].legend()
plt.show()
```

Prueba de  Dickey Fuller


\begin{cases}
H_0: z_t \  \text{es  no  estacionaria}\\
H_1: z_t \ \text{es estacionaria}
\end{cases}

rechazar $H_0  \ \text{ si p-value} < \alpha$

```python
# Prueba de ra√≠z u nitaria (ADF)
def adf_test(serie, nombre):
    result = adfuller(serie, autolag='BIC')
    print(f"ADF Test para {nombre}:")
    print(f"  Estad√≠stico: {result[0]:.4f}, p-valor: {result[1]:.4f}")
    print("  ¬øEstacionaria?:", "S√≠" if result[1] < 0.05 else "No")
    print("-" * 50)

```

```python

adf_test(une, "Desempleo (U)")
adf_test(gdp, "PIB (GDP)")
```

```python
# Crear dataset bivariado
data_bv = datos[['GDP', 'U']].dropna()
```

```python

# Selecci√≥n del rezago √≥ptimo
modelo_var = VAR(data_bv)
resumen = modelo_var.select_order(maxlags=12)
print("Criterios de informaci√≥n:")
print(resumen.summary())
```

```python
# Estimaci√≥n del modelo VAR con 2 rezagos
modelo = modelo_var.fit(2)
print(modelo.summary())
```

```python
# Prueba de estabilidad (ra√≠ces dentro del c√≠rculo unitario)
print("Ra√≠ces del modelo VAR:")
print(modelo.roots)
```

```python
abs(modelo.roots)
```

```python
modelo.resid
```

###  Aplica Ljung-Box a cada serie de residuos

```python

from statsmodels.stats.diagnostic import acorr_ljungbox

resid = modelo.resid


for col in resid.columns:
    lb_test = acorr_ljungbox(resid[col], lags=[8], return_df=True)
    print(f"{col}: p-value = {lb_test['lb_pvalue'].values[0]}")

```

### Normalidad

```python
#  Normalidad (Jarque-Bera)
print("\nTest de normalidad (Jarque-Bera):")
for col in modelo.resid.columns:
    jb = jarque_bera(modelo.resid[col])
    print(f"{col}: p-valor = {jb.pvalue:.4f}")
```

```python
print(modelo.is_stable())  # True si cumple el supuesto
```

###  Heterocedasticidad (ARCH test)

```python

print("\nTest ARCH (heterocedasticidad):")
for col in modelo.resid.columns:
    print(f"{col}:")
    arch_result = het_arch(modelo.resid[col])
    print(f"  p-valor: {arch_result[1]:.4f}")
```

###  Causalidad de Granger

```python

print("\nCausalidad de Granger:")
print("GDP causa UNE:")
print(modelo.test_causality(causing='GDP', caused='U', kind='f'))
print("UNE causa GDP:")
print(modelo.test_causality(causing='U', caused='GDP', kind='f'))
```

# Funciones de Impulso-Respuesta (IRF)

üìò ¬øQu√© es una Funci√≥n de Impulso-Respuesta (IRF)?

La Funci√≥n de Impulso-Respuesta (IRF, por sus siglas en ingl√©s: Impulse Response Function) es una herramienta fundamental en el an√°lisis de modelos VAR (Vector Autoregressive), que permite estudiar c√≥mo una variable responde a un shock (perturbaci√≥n inesperada) en otra variable del sistema a lo largo del tiempo.

üß† Interpretaci√≥n intuitiva:
La IRF muestra c√≥mo cambia una variable dependiente tras un shock unitario en una variable explicativa, manteniendo todas las dem√°s condiciones constantes.

‚öôÔ∏è ¬øQu√© mide exactamente?
El impacto din√°mico de una perturbaci√≥n en una variable sobre s√≠ misma o sobre otras variables del modelo.

La respuesta se mide durante varios per√≠odos futuros (por ejemplo, d√≠as, meses o trimestres).

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

irf = modelo.irf(20)

fig = irf.plot(impulse='U', response='GDP')
fig.set_size_inches(6, 4)  
plt.title("IRF: Impacto de U sobre GDP")
plt.tight_layout()
plt.show()
```

üß† Lectura de la curva:
- En el periodo 0, el impacto inicial del shock en el desempleo sobre el PIB es negativo, indicando una ca√≠da inmediata del PIB.

- En los  periodos (1 a 6), el PIB responde de forma positiva, alcanzando un m√°ximo alrededor del periodo 5, lo que podr√≠a interpretarse como una correcci√≥n del efecto inicial.

- Despu√©s del pico, el efecto comienza a disminuir gradualmente, acerc√°ndose a cero, lo que indica que el impacto del shock en U sobre GDP es transitorio.

- Las l√≠neas punteadas representan los intervalos de confianza 95%. 

```python
import matplotlib.pyplot as plt

irf = modelo.irf(20)
fig = irf.plot(impulse='GDP', response='U')  
fig.set_size_inches(6, 4)  # Cambia el tama√±o
fig.suptitle("IRF: Impacto de GDP sobre U")  # 
plt.tight_layout()
plt.show()

```

- Luego, el efecto empieza a reducirse en magnitud, y para el periodo 12 en adelante, la tasa de desempleo vuelve gradualmente a su nivel base, acerc√°ndose a cero.

# Descomposici√≥n de la varianza

Qu√© es la Descomposici√≥n de la Varianza (FEVD)?

La Descomposici√≥n de la Varianza del Error de Pron√≥stico (Forecast Error Variance Decomposition, FEVD) permite cuantificar qu√© proporci√≥n de la incertidumbre (varianza) en el pron√≥stico de una variable se debe a los shocks propios y a los shocks de las dem√°s variables del modelo VAR.

üìà ¬øQu√© muestra?
Por ejemplo, si estamos pronosticando el PIB (GDP), el FEVD indica cu√°nto de su variabilidad futura se explica por shocks en el PIB mismo, en el desempleo (U), u otras variables del sistema.

El an√°lisis se realiza para distintos horizontes de tiempo (por ejemplo, 1, 5, 10 periodos hacia adelante).

```python
fevd = modelo.fevd(10)
fevd.plot()
plt.show()
```

- En los primeros periodos, casi el 100% de la variabilidad del pron√≥stico del PIB se explica por shocks en el mismo PIB.

A medida que pasan los periodos, una peque√±a proporci√≥n (alrededor del 10%) comienza a explicarse por shocks en el desempleo (U).

Esto indica que el PIB es fuertemente autoexplicativo, y los shocks en la tasa de desempleo tienen un efecto limitado pero creciente sobre su incertidumbre futura.

- En el periodo 0, m√°s del 50% de la incertidumbre sobre el desempleo se explica por shocks externos (principalmente del PIB).

- A medida que transcurren los periodos, los shocks propios de GDP (barras negras) ganan relevancia y explican cerca del 70% de la variabilidad futura.

- Esto sugiere que, en el corto plazo (2 periodos), el desempleo es influenciado por el mismo y  otras variables del sistema (como el PIB) 50-50% , pero con el tiempo por shocks propios(30%) y  externos (70%).

```python

# Pron√≥stico (8 trimestres)
forecast = modelo.forecast_interval(data_bv.values[-2:], steps=8)
forecast_df = pd.DataFrame(forecast[0], columns=['GDP', 'U'])
forecast_df.index = pd.date_range(start=datos.index[-1] + pd.offsets.QuarterEnd(), periods=8, freq='Q')
forecast_df.plot(title="Pron√≥stico de GDP y U")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Hacer pron√≥stico con intervalos de confianza
forecast_mean, lower, upper = modelo.forecast_interval(data_bv.values[-modelo.k_ar:], steps=8)

# 2. Usar solo las columnas que entrenaron el modelo
cols = data_bv.columns  # ‚Üê t√≠picamente ['GDP', 'U']

# 3. Crear DataFrames para pron√≥stico e intervalos
forecast_df = pd.DataFrame(forecast_mean, columns=cols)
lower_df = pd.DataFrame(lower, columns=cols)
upper_df = pd.DataFrame(upper, columns=cols)

# 4. Crear √≠ndice de fechas para el pron√≥stico
forecast_index = pd.date_range(start=datos.index[-1] + pd.offsets.QuarterEnd(), periods=8, freq='Q')
forecast_df.index = lower_df.index = upper_df.index = forecast_index

# 5. Ajuste del modelo (valores ajustados del VAR)
fitted_df = pd.DataFrame(modelo.fittedvalues, index=datos.index[modelo.k_ar:], columns=cols)

# 6. Crear gr√°ficos
fig, axes = plt.subplots(2, 1, figsize=(10, 7), sharex=True)

for i, var in enumerate(cols):
    # Serie observada
    axes[i].plot(datos.index, datos[var], label='Observado', color='black')

    # Ajuste del modelo (in-sample)
    axes[i].plot(fitted_df.index, fitted_df[var], label='Ajuste VAR', color='blue', linestyle='dotted')

    # Pron√≥stico
    axes[i].plot(forecast_df.index, forecast_df[var], label='Pron√≥stico', color='orange', linestyle='--')

    # Intervalo de confianza
    axes[i].fill_between(forecast_df.index, lower_df[var], upper_df[var],
                         color='orange', alpha=0.2, label='IC 95%')

    axes[i].set_title(f'Variable: {var}')
    axes[i].legend()

# T√≠tulo general
plt.suptitle("Pron√≥stico con Ajuste VAR e Intervalos de Confianza", fontsize=14)
plt.tight_layout()
plt.show()
```
