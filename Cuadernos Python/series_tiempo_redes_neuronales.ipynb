## **Redes Neuronales Recurrentes (RNN)**


Las **Redes Neuronales Recurrentes (RNN)** son un tipo de redes diseñadas para trabajar con **datos secuenciales**, donde el orden de los datos es fundamental. A diferencia de las redes neuronales tradicionales (feedforward), las RNN **recuerdan información previa** mediante un estado interno o "memoria".

## ¿Cómo funcionan las RNN?

Cada paso en la secuencia actualiza un **estado oculto (memoria)**. La **unidad básica** es la **célula recurrente**, que recibe la entrada actual y el estado oculto anterior. Con esta información, produce una salida y un nuevo estado oculto que se utiliza en el siguiente paso:



Este mecanismo permite que la red tenga **"memoria" del pasado** al procesar la secuencia.

#### Limitaciones de las RNN tradicionales

Aunque las RNN pueden aprender patrones en secuencias, presentan dificultades cuando deben capturar **dependencias a largo plazo**. Esto se debe a problemas como el **desvanecimiento o explosión del gradiente** durante el entrenamiento.

#### LSTM y GRU: Variantes mejoradas

Para superar estas limitaciones, se desarrollaron variantes como:

- **🧠 LSTM (Long Short-Term Memory):**
  - Usa tres puertas: entrada, olvido y salida.
  - Retiene información relevante durante muchos pasos.
  - Ideal para secuencias largas y complejas.

- **⚙️ GRU (Gated Recurrent Unit):**
  - Más simple y rápida que LSTM.
  - Usa dos puertas: actualización y reinicio.
  - Buen rendimiento con menor complejidad computacional.

#### Aplicaciones comunes

- Predicción de series temporales (clima, ventas, salud).
- Modelado de lenguaje natural (traducción automática, chatbots).
- Reconocimiento de voz.
- Análisis de secuencias biológicas.

---

> En resumen, las RNN permiten que una red neuronal trabaje con secuencias gracias a su capacidad de memoria. Sin embargo, para problemas complejos con dependencias a largo plazo, las arquitecturas **LSTM** y **GRU** son las más recomendadas.

### 🧠 Long Short-Term Memory (LSTM)

Las **redes neuronales LSTM (Long Short-Term Memory)** son un tipo especializado de RNN (Red Neuronal Recurrente) diseñadas para superar las limitaciones de las RNN tradicionales en la captura de **dependencias temporales a largo plazo**.

A diferencia de las RNN simples, las LSTM incorporan una arquitectura más compleja que incluye **unidades de memoria internas** y **mecanismos de puertas** (gates) que controlan de manera precisa el flujo de información a lo largo del tiempo.


### 🔧 Estructura de una celda LSTM

Una celda LSTM contiene **tres puertas principales** que permiten decidir qué información se mantiene, se actualiza o se elimina:

#### 🔹 Puerta de Olvido (Forget Gate)
- Decide **qué parte de la información pasada se olvida**.
- Usa la entrada actual (`xₜ`) y la salida anterior (`hₜ₋₁`) con una función sigmoide.
- Resultado entre 0 (olvidar todo) y 1 (mantener todo).

#### 🔹 Puerta de Entrada (Input Gate)
- Determina **qué nueva información debe almacenarse** en la memoria.
- Combina la entrada actual y la salida anterior para decidir cuánto de la nueva información se guarda.

#### 🔹 Puerta de Salida (Output Gate)
- Decide **cuánta información de la memoria se envía como salida actual**.
- Regula qué parte del estado interno se convierte en salida hacia la siguiente celda.

## Tipos de Problemas en el Modelado de Series Temporales

Cuando se trabaja con modelos de series temporales, es importante tener en cuenta tres preguntas clave que definen la complejidad del problema:

1. ¿Qué serie o series se van a usar para **entrenar** el modelo?
2. ¿Qué serie o series se desean **predecir**?
3. ¿Cuántos pasos hacia el futuro se quieren **pronosticar**?

Responder estas preguntas nos ayuda a identificar el tipo de problema y elegir la arquitectura de red neuronal más adecuada.

---

## 🤖 ¿Cómo ayudan las RNN?

Las **Redes Neuronales Recurrentes (RNN)**, gracias a su flexibilidad, pueden adaptarse a diferentes escenarios de modelado de series temporales:

---

### 🔹 **Problema 1:1 — Una serie de entrada, una serie de salida**

- **Nombre técnico**: Single-series, single-output  
- **Descripción**: Se utiliza el pasado de una única serie para predecir su propio valor futuro.
- **Ejemplo**: Predecir la temperatura de mañana usando las temperaturas de los últimos 7 días.

---

### 🔹 **Problema N:1 — Varias series de entrada, una serie de salida**

- **Nombre técnico**: Multi-series, single-output  
- **Descripción**: Se usan varias series (por ejemplo, variables meteorológicas) para predecir una sola variable de interés.
- **Ejemplo**: Predecir la temperatura usando temperatura, humedad y presión atmosférica de días anteriores.

---

### 🔹 **Problema N:M — Varias series de entrada, varias series de salida**

- **Nombre técnico**: Multi-series, multiple-outputs  
- **Descripción**: Se desea pronosticar el futuro de varias series al mismo tiempo usando información pasada de todas ellas.
- **Ejemplo**: Predecir los precios futuros de varias acciones de bolsa usando sus históricos, junto con datos de energía y materias primas.

---

## 🕒 ¿Predicción a corto o largo plazo?

En cualquiera de los casos anteriores, se puede hacer:

- **Single-step forecasting**: Se predice **un único valor futuro** (por ejemplo, la temperatura de mañana).
- **Multi-step forecasting**: Se predicen **varios pasos hacia el futuro** (por ejemplo, la temperatura de los próximos 5 días).

## 📊 ¿Por qué dividir en Train, Validación y Test?

Cuando se trabaja con series temporales, es importante separar los datos en diferentes segmentos para garantizar un buen desempeño del modelo y una evaluación justa. Generalmente se dividen en:

---

### 🧱  Conjunto de Entrenamiento (**Train**)
- Contiene los datos históricos con los que el modelo **aprende los patrones**.
- Aquí es donde se ajustan los pesos internos del modelo (por ejemplo, de una red neuronal).
- Debe ser lo más largo posible para que el modelo vea ciclos, estacionalidades, etc.

---

### 🧪 Conjunto de Validación (**Validation**)
- Se usa para **probar el modelo durante el entrenamiento**, sin usar los datos futuros.
- Sirve para:
  - Ajustar **hiperparámetros** (número de pasos, neuronas, regularización, etc.).
  - Evitar **sobreajuste** (overfitting).
  - Elegir el mejor momento para detener el entrenamiento (**early stopping**).
- Es **clave cuando usas redes neuronales o modelos complejos**.

---

### 🧮  Conjunto de Prueba (**Test**)
- Se usa **al final** para evaluar el modelo **como si predijera el futuro real**.
- No debe tocarse durante el entrenamiento.
- Refleja el rendimiento del modelo en condiciones reales.

---

### ❓ ¿Por qué **no es suficiente** usar solo Train y Test?

Usar solo `train` y `test` puede llevar a:

- Elegir un modelo que **funciona bien por suerte** en el test, pero no generaliza.
- **Ajustar hiperparámetros sobre el test**, contaminando la evaluación final.
- Mayor riesgo de **sobreajuste**.

La validación es un paso intermedio que permite encontrar un buen modelo **sin mirar los datos futuros**.

---

### 🧠 Analogía

- **Entrenamiento**: Es como estudiar para un examen.
- **Validación**: Hacer simulacros o ensayos antes del examen real.
- **Prueba**: El examen final, sin ayuda ni ajustes.

---

### ✅ Recomendación práctica

- Usa `train` largo (por ejemplo, 70% del tiempo).
- Usa `validation` reciente (15–20%) para seleccionar y ajustar.
- Usa `test` final (último 10–15%) para evaluar rendimiento real.

````python
# ========================================
# MODELO LSTM - FLUJO COMPLETO (O3)
# ========================================

# 1. Selección de variable y escalado
# ------------------------------------------------------------------------------
from sklearn.preprocessing import MinMaxScaler

series = 'o3'         # Variable a modelar
lags = 24             # Ventana temporal (ej. 24 horas)
steps = 1             # Pasos a futuro (solo 1 en este caso)
````

📌 Aquí definimos los parámetros del modelo:

series indica la variable objetivo que queremos predecir.

lags = 24 quiere decir que vamos a usar las 24 horas anteriores como entrada para predecir.

steps = 1 indica que queremos predecir el valor de la siguiente hora solamente.

````python

data_train = air_quality_train[[series]].copy()
data_val   = air_quality_val[[series]].copy()
data_test  = air_quality_test[[series]].copy()
````

📌 Aquí tomamos solo la variable 'o3' desde las particiones de datos:

data_train: datos de entrenamiento (usados para aprender).

data_val: datos de validación (para ajustar hiperparámetros).

data_test: datos de prueba final (para evaluar el modelo).

Usamos [[series]] para mantener el formato de DataFrame.

### 🔢 ¿Qué es el `MinMaxScaler`?

El `MinMaxScaler` es una técnica de **escalado de datos** utilizada en Machine Learning y Deep Learning para transformar valores numéricos dentro de un **rango específico**, normalmente entre 0 y 1.

Esto es especialmente útil cuando se entrenan redes neuronales como LSTM, ya que estos modelos funcionan mejor cuando las entradas están en la misma escala.


#### 📐 Fórmula del escalado

Para cada valor $x$, el escalado se calcula así:

$$
x_{\text{escalado}} = \frac{x - x_{\text{min}}}{x_{\text{max}} - x_{\text{min}}}
$$

Donde:
- $ x$: valor original
- $ x_{\text{min}}$: valor mínimo de la variable
- $x_{\text{max}}$: valor máximo de la variable




````python

# Escalado de los datos
scaler = MinMaxScaler()
data_train_scaled = scaler.fit_transform(data_train)
data_val_scaled   = scaler.transform(data_val)
data_test_scaled  = scaler.transform(data_test)

````

Este bloque aplica escalado normalizado para que todos los valores estén entre 0 y 1, lo cual es importante para que el modelo LSTM pueda aprender mejor.

Se ajusta (fit) y transforma solo con los datos de entrenamiento.

Luego se transforma validación y prueba usando los mismos límites.

🎯 Esto evita fuga de información del futuro hacia el pasado y garantiza consistencia.



````python
# Convertir el array escalado en DataFrame con las fechas originales
data_train_scaled_df = pd.DataFrame(
    data_train_scaled,
    index=data_train.index,  # ← aquí se recuperan las fechas
    columns=[series]         # asegura nombre de columna
)

# Gráfico con fechas en el eje X
plt.figure(figsize=(12, 4))
plt.plot(data_train_scaled_df, label='Train Escalado', color='blue')
plt.title('Serie Escalada (MinMaxScaler)')
plt.xlabel('Fecha')
plt.ylabel('Valor Escalado')
plt.grid(True)
plt.legend()
plt.show()
````

````python

# 2. Crear secuencias
# ------------------------------------------------------------------------------
import numpy as np

def crear_secuencias(data, lags):
    X, y = [], []
    for i in range(len(data) - lags):
        X.append(data[i:i+lags])
        y.append(data[i+lags])
    return np.array(X), np.array(y)

X_train, y_train = crear_secuencias(data_train_scaled, lags)
X_val, y_val     = crear_secuencias(data_val_scaled, lags)

````

## 🧩 ¿Para qué sirve la función `crear_secuencias`?

Cuando trabajamos con series temporales (por ejemplo, niveles de contaminación por hora), no podemos darle los datos "en bruto" al modelo. Necesitamos organizar la información para que el modelo aprenda **usando el pasado para predecir el futuro**.



### 🧠 ¿Qué hace esta función?

La función `crear_secuencias` transforma una serie de tiempo en muchas pequeñas "ventanas de datos".  
Cada una de esas ventanas contiene un número fijo de valores pasados (por ejemplo, las últimas 24 horas), y su objetivo es **predecir el valor que viene después**.


### 🎯 ¿Por qué es importante?

- Permite que el modelo LSTM entienda los **patrones a lo largo del tiempo**.
- Convierte una sola secuencia larga en muchos ejemplos para entrenar el modelo.
- Prepara los datos de forma que el modelo pueda hacer predicciones paso a paso.


### 📌 Ejemplo sencillo

Si tuvieras los datos `[10, 12, 14, 16, 18]` y usas ventanas de 3 valores:

- El modelo ve `[10, 12, 14]` → y aprende a predecir `16`
- Luego ve `[12, 14, 16]` → y aprende a predecir `18`

Así se entrena paso a paso.


### ✅ En resumen:

> "La función `crear_secuencias` prepara los datos para que el modelo pueda aprender del pasado y hacer predicciones del futuro."

````python

# 3. Redimensionar para LSTM
# ------------------------------------------------------------------------------
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_val   = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))

````

#### 📐 Redimensionar para LSTM

Las redes LSTM necesitan que los datos tengan 3 dimensiones:

[samples, timesteps, features]


Este paso convierte cada secuencia en ese formato.  
Como solo usamos una variable (`o3`), se añade una dimensión extra con `features = 1`.

> Así el modelo entiende que cada entrada es una secuencia de 24 pasos con 1 dato por paso.

````python

# 4. Definir modelo LSTM
# ------------------------------------------------------------------------------
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.optimizers import Adam
from keras.losses import MeanSquaredError
from keras.callbacks import EarlyStopping

model = Sequential()
model.add(LSTM(units=64, input_shape=(lags, 1)))
model.add(Dense(1))  # predicción de un paso
model.compile(optimizer=Adam(learning_rate=0.01), loss=MeanSquaredError())

model.summary()

````

## ⚙️ Definición del modelo LSTM

En este modelo usamos una red neuronal secuencial (`Sequential`) con dos capas principales:

1. **Capa LSTM**:  
   - Tiene 64 unidades (neuronas).
   - Aprende patrones a partir de la secuencia de entrada (lags).
   - Es capaz de captar relaciones temporales entre los datos.

2. **Capa Densa (Dense)**:  
   - Toma la salida final de la LSTM y la convierte en una única predicción (1 valor futuro).

---

### 🔧 Configuración del modelo

El modelo se compila con:

- **Optimizador**: `Adam` (con `learning_rate=0.01`), recomendado para deep learning.
- **Función de pérdida**: `MeanSquaredError` (error cuadrático medio), ideal para regresión.

---

### 🧾 Resumen del modelo

| Capa         | Tipo   | Salida (Output Shape) | Parámetros (#) |
|--------------|--------|------------------------|-----------------|
| LSTM         | LSTM   | `(None, 64)`           | 16,896          |
| Capa Densa   | Dense  | `(None, 1)`            | 65              |
| **Total**    |        |                        | **16,961**      |

📌 Todos los parámetros son entrenables.

---

> ✅ Este modelo toma una secuencia de valores del pasado (por ejemplo, 24 horas) y predice el siguiente valor. Es simple pero muy efectivo para problemas de series temporales.


````python

# 5. Entrenamiento
# ------------------------------------------------------------------------------
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10,
    batch_size=32,
    callbacks=[EarlyStopping(monitor="val_loss", patience=5)],
    verbose=1
)
````

## 🏋️‍♂️ Entrenamiento del modelo

El modelo se entrena con los siguientes parámetros:

- `epochs = 10`: número máximo de veces que se recorre todo el conjunto de entrenamiento.
- `batch_size = 32`: cantidad de muestras procesadas antes de actualizar los pesos.
- `validation_data`: se evalúa el rendimiento con datos de validación en cada época.
- `EarlyStopping`: detiene el entrenamiento si no mejora la pérdida de validación por 5 épocas seguidas (`patience=5`).

````python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler

def predecir_futuro(model, data_esc, lags, pasos_a_futuro, scaler, index_original):
    """
    Realiza predicción fuera de muestra paso a paso con un modelo LSTM univariado.
    
    Parámetros:
    - model: modelo LSTM ya entrenado
    - data_esc: datos escalados (array 2D o DataFrame) de la variable objetivo
    - lags: número de pasos anteriores que usa el modelo
    - pasos_a_futuro: cantidad de pasos a predecir
    - scaler: objeto MinMaxScaler ajustado con train
    - index_original: índice temporal del DataFrame original para generar fechas futuras

    Devuelve:
    - DataFrame con fechas futuras y predicciones desescaladas
    """

    # 1. Extraer última secuencia de entrada conocida
    última_secuencia = data_esc[-lags:].reshape(1, lags, 1)
    secuencia_actual = última_secuencia.copy()

    # 2. Generar predicciones en cadena
    predicciones = []
    for _ in range(pasos_a_futuro):
        pred = model.predict(secuencia_actual, verbose=0)[0][0]
        predicciones.append(pred)
        nueva_secuencia = np.append(secuencia_actual[0, 1:, 0], pred).reshape(1, lags, 1)
        secuencia_actual = nueva_secuencia

    # 3. Desescalar predicciones
    predicciones_desescaladas = scaler.inverse_transform(np.array(predicciones).reshape(-1, 1)).flatten()

    # 4. Generar fechas futuras
    última_fecha = index_original[-1]
    frecuencia = pd.infer_freq(index_original)
    if frecuencia is None:
        frecuencia = 'H'  # frecuencia por defecto (horaria)
    fechas_futuras = pd.date_range(
        start=última_fecha + pd.to_timedelta(f'1{frecuencia}'),
        periods=pasos_a_futuro,
        freq=frecuencia
    )

    # 5. Crear DataFrame con resultados
    df_forecast = pd.DataFrame({
        'Fecha': fechas_futuras,
        'Pronóstico_O3': predicciones_desescaladas
    }).set_index('Fecha')

    return df_forecast

# Ejemplo de uso:
df_forecast = predecir_futuro(
    model=model,
    data_esc=data_test_scaled,
    lags=24,
    pasos_a_futuro=24,
    scaler=scaler,
    index_original=data_test.index
)

px.line(df_forecast, y='Pronóstico_O3', title='Pronóstico fuera de muestra de O₃ (24 horas)').show()
````

## Multi-step forecasting

En este caso, se desea predecir los próximos 5 valores de O3 utilizando únicamente sus datos históricos. Se trata por lo tanto de un escenario en el que múltiples pasos a futuro de una única serie temporal se modela utilizando únicamente sus valores pasados.

Para ello se utilizará una arquitectura similar a la anterior, pero con un mayor número de neuronas en la capa LSTM y en la primera capa densa. Esto permitirá al modelo tener mayor flexibilidad para modelar la serie temporal.

````python
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.optimizers import Adam
from keras.losses import MeanSquaredError
from keras.callbacks import EarlyStopping

def crear_secuencias_multistep(data, lags, steps):
    X, y = [], []
    for i in range(len(data) - lags - steps + 1):
        X.append(data[i:i+lags])
        y.append(data[i+lags:i+lags+steps].flatten())
    return np.array(X), np.array(y)

from sklearn.metrics import mean_squared_error

def entrenar_y_predecir_multistep(data_train_scaled, data_val_scaled, data_test_scaled, lags, steps, scaler):
    # Crear secuencias
    X_train, y_train = crear_secuencias_multistep(data_train_scaled, lags, steps)
    X_val, y_val     = crear_secuencias_multistep(data_val_scaled, lags, steps)
    X_test, y_test   = crear_secuencias_multistep(data_test_scaled, lags, steps)

    # Redimensionar
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_val   = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))
    X_test  = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

    # Crear modelo
    model = Sequential()
    model.add(LSTM(64, input_shape=(lags, 1)))
    model.add(Dense(steps))
    model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')

    # Entrenar
    model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=15,
        batch_size=32,
        callbacks=[EarlyStopping(monitor="val_loss", patience=5)],
        verbose=0
    )

    # Predecir
    y_pred_scaled = model.predict(X_test)

    # Desescalar
    y_test_inv = np.array([scaler.inverse_transform(y.reshape(-1, 1)).flatten() for y in y_test])
    y_pred_inv = np.array([scaler.inverse_transform(y.reshape(-1, 1)).flatten() for y in y_pred_scaled])

    # Calcular MSE promedio por muestra
    mse_total = mean_squared_error(y_test_inv.flatten(), y_pred_inv.flatten())

    return y_test_inv, y_pred_inv, mse_total




````

````python
steps_list = [24, 48, 72, 120]
lags = 48

mse_resultados = {}

fig, axes = plt.subplots(len(steps_list), 1, figsize=(10, 3 * len(steps_list)))

for i, steps in enumerate(steps_list):
    y_test_inv, y_pred_inv, mse_total = entrenar_y_predecir_multistep(
        data_train_scaled, data_val_scaled, data_test_scaled, lags, steps, scaler
    )

    mse_resultados[steps] = mse_total

    axes[i].plot(y_test_inv[0], label='Real', color='black')
    axes[i].plot(y_pred_inv[0], label='Predicho', color='red', linestyle='--')
    axes[i].set_title(f'Predicción multistep ({steps} pasos) - MSE: {mse_total:.2f}')
    axes[i].set_ylabel('O₃')
    axes[i].grid(True)
    axes[i].legend()

axes[-1].set_xlabel('Paso en el horizonte')
plt.tight_layout()
plt.show()
````

## Problemas N:1 - Series temporales múltiples con salida única¶

En este caso se tratará de predecir la misma serie temporal, pero utilizando múltiples series temporales como predictores. Se trata, por lo tanto, de un escenario en el que valores pasados de múltiples series temporales se utilizan predecir una única serie temporal.

Este tipo de aproximaciones son muy útiles cuando se dispone de múltiples series temporales relacionadas entre sí. Por ejemplo, en el caso de la predicción de la temperatura, se pueden utilizar múltiples series temporales como la humedad, la presión atmosférica, la velocidad del viento, etc.

En este tipo de problemas, la arquitectura de la red neuronal es más compleja, se necesita una capa densa recurrente adicional para procesar las múltiples series de entrada. Además, se añade otra capa oculta densa para procesar la salida de la capa recurrente. Como se puede observar, la creación del modelo utilizando skforecast es muy sencilla, simplemente basta con pasar una lista de enteros al arguento recurrent_units y dense_units para crear múltiples capas recurrentes y densas.

````python
# ================================
# 1. Configuración inicial
# ================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.optimizers import Adam
from keras.losses import MeanSquaredError
from keras.callbacks import EarlyStopping

# ================================
# 2. Parámetros
# ================================
series = ['pm2.5', 'co', 'no', 'no2', 'pm10', 'nox', 'o3', 'veloc.', 'direc.','so2']
target = 'o3'
lags = 32
steps = 5

# ================================
# 3. División train, val, test
# ================================
air_quality_full = air_quality.copy()  # debe tener todas las columnas

air_quality_train = air_quality_full.loc[:'2023-03-30 23:00:00', series].copy()
air_quality_val   = air_quality_full.loc['2023-03-30 23:00:00':'2023-09-30 23:00:00', series].copy()
air_quality_test  = air_quality_full.loc['2023-09-30 23:00:00':, series].copy()

# ================================
# 4. Escalado por variable
# ================================
scalers = {}
for col in series:
    scaler = MinMaxScaler()
    air_quality_train[col] = scaler.fit_transform(air_quality_train[[col]])
    air_quality_val[col]   = scaler.transform(air_quality_val[[col]])
    air_quality_test[col]  = scaler.transform(air_quality_test[[col]])
    scalers[col] = scaler  # guardar para desescalar

# ================================
# 5. Crear secuencias multistep
# ================================
def crear_secuencias_multiserie_multistep(data, target, lags, steps):
    X, y = [], []
    for i in range(len(data) - lags - steps + 1):
        seq_x = data.iloc[i:i+lags][series].values
        seq_y = data.iloc[i+lags:i+lags+steps][target].values
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

X_train, y_train = crear_secuencias_multiserie_multistep(air_quality_train, target, lags, steps)
X_val, y_val     = crear_secuencias_multiserie_multistep(air_quality_val, target, lags, steps)
X_test, y_test   = crear_secuencias_multiserie_multistep(air_quality_test, target, lags, steps)

# ================================
# 6. Redimensionar para LSTM
# ================================
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], len(series)))
X_val   = X_val.reshape((X_val.shape[0], X_val.shape[1], len(series)))
X_test  = X_test.reshape((X_test.shape[0], X_test.shape[1], len(series)))

# ================================
# 7. Crear modelo
# ================================
model = Sequential()
model.add(LSTM(100, return_sequences=True, input_shape=(lags, len(series))))
model.add(LSTM(50))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(steps))
model.compile(optimizer=Adam(learning_rate=0.01), loss=MeanSquaredError())
model.summary()

# ================================
# 8. Entrenamiento
# ================================
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=20,
    batch_size=32,
    callbacks=[EarlyStopping(monitor="val_loss", patience=5)],
    verbose=1
)

# ================================
# 9. Predicción y desescalado
# ================================
y_pred_scaled = model.predict(X_test)

scaler_target = scalers[target]
y_test_inv = np.array([scaler_target.inverse_transform(y.reshape(-1, 1)).flatten() for y in y_test])
y_pred_inv = np.array([scaler_target.inverse_transform(y.reshape(-1, 1)).flatten() for y in y_pred_scaled])

# ================================
# 10. Visualización ejemplo
# ================================
plt.figure(figsize=(10, 4))
plt.plot(y_test_inv[0], label='Real', color='black')
plt.plot(y_pred_inv[0], label='Predicho', color='red', linestyle='--')
plt.title(f'Predicción multiserie a 5 pasos - Variable: {target}')
plt.xlabel('Paso en el horizonte')
plt.ylabel(f'Concentración de {target.upper()}')
plt.grid(True)
plt.legend()
plt.show()
````
