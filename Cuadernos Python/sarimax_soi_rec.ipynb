## **Introducción Series de Tiempo Multivariadas**
Las series de tiempo multivariadas (MTS, por sus siglas en inglés) son una extensión del análisis de series de tiempo que permite observar y modelar el comportamiento de múltiples variables simultáneamente a lo largo del tiempo. En lugar de enfocarse en una única variable, este enfoque se preocupa por cómo un conjunto de variables cambia, interactúa y coevoluciona, lo que permite capturar la complejidad y la interdependencia en sistemas reales.





**Definición de las Series de Tiempo Multivariadas**

Una serie de tiempo multivariada se compone de observaciones de más de una variable registradas en intervalos de tiempo regulares. Por ejemplo, en un sistema económico, podríamos observar la evolución simultánea del Producto Interno Bruto (PIB), la tasa de inflación, y el desempleo durante varios años, y analizar cómo estas variables se afectan mutuamente.




**Comparación con las Series de Tiempo Univariadas**

En un análisis univariado, se modela y predice una sola variable basándose en sus valores pasados. Ejemplo: Predecir la temperatura diaria de una ciudad basándose en sus registros históricos.
En el análisis multivariado, se modelan múltiples variables a la vez, considerando sus relaciones e influencias mutuas. Ejemplo: Predecir la temperatura, la humedad, y la velocidad del viento simultáneamente.



**Importancia y Ventajas de las Series de Tiempo Multivariadas**

Captura de Interdependencias: Permite entender cómo diferentes variables influyen entre sí, proporcionando un análisis más realista y completo. Por ejemplo, en un sistema financiero, el precio de una acción puede depender de la tasa de interés, la inflación y otros factores económicos.

Mejor capacidad de predicción: Al incorporar múltiples variables, los modelos multivariados suelen ser más precisos que los univariados, ya que aprovechan información adicional que las variables externas proporcionan.
Análisis de Causalidad: Facilita el análisis de relaciones causa-efecto entre variables, ayudando a identificar qué factores son impulsores clave dentro del sistema.



**Ejemplos Prácticos de Series de Tiempo Multivariadas**

- Finanzas: Análisis simultáneo de precios de acciones, tasas de interés y volúmenes de transacción para identificar tendencias del mercado.
- Economía: Monitoreo conjunto del PIB, la tasa de desempleo y la inflación para evaluar la salud económica de un país.
- Medicina: Seguimiento de múltiples signos vitales (frecuencia cardíaca, presión arterial, niveles de oxígeno) para evaluar la condición de un paciente en tiempo real.
- Climatología: Estudio conjunto de la temperatura, la humedad, la precipitación y la velocidad del viento para comprender los patrones climáticos.




**Principales Modelos para Series de Tiempo Multivariadas**

- Modelo VAR (Vector Autoregressive)
Cada variable se modela como una combinación lineal de sus propios valores pasados y los valores pasados de las demás variables.
Ideal para datos donde las variables influyen mutuamente sin una estructura jerárquica definida.

- Modelo VECM (Vector Error Correction Model)
Variante del VAR para series cointegradas, es decir, variables que tienen una relación de equilibrio a largo plazo.
Útil cuando las variables presentan una tendencia común a largo plazo, pero muestran fluctuaciones a corto plazo.

- Modelo VARMA (Vector Autoregressive Moving Average)
Combina componentes autoregresivos y de promedio móvil en un modelo multivariado.
Captura tanto la influencia de los valores pasados como de los errores pasados.

- Modelos con Variables Exógenas (VARX y VARMAX)
Incluyen variables exógenas que afectan el sistema, es decir, factores externos que influyen en las variables del modelo pero no son influenciados por ellas.




**Análisis y Desafíos en Series de Tiempo Multivariadas**

- Cointegración:
En series de tiempo multivariadas, es común encontrar variables que se mueven juntas a largo plazo. El análisis de cointegración identifica y modela estas relaciones, crucial para evitar conclusiones erróneas.

- Causalidad de Granger:
Determina si una variable es útil para predecir otra. Por ejemplo, si los datos históricos de la inflación pueden ayudar a predecir el PIB, decimos que la inflación "causa" el PIB en el sentido de Granger.

- Alta Dimensionalidad:
El número de parámetros a estimar crece rápidamente con el número de variables y rezagos considerados. Técnicas como la reducción de dimensionalidad (PCA) y la regularización (LASSO) son útiles para abordar este problema.

- Estabilidad del Modelo:
Los modelos multivariados pueden volverse inestables si las variables no son estacionarias. La diferenciación o la cointegración son herramientas comunes para abordar esto.



**Aplicaciones en el Mundo Real**

- Finanzas: Análisis de portafolios, donde múltiples activos se modelan conjuntamente para gestionar riesgos y maximizar retornos.
- Economía: Evaluación del impacto de políticas monetarias en el crecimiento económico, el desempleo y la inflación.
- Industria y Manufactura: Monitoreo de múltiples procesos de producción para identificar cuellos de botella y mejorar la eficiencia.
- Climatología y Medio Ambiente: Modelos para predecir patrones climáticos y evaluar el impacto del cambio climático.



**Herramientas y Software para el Análisis de Series de Tiempo Multivariadas**

- En Python: statsmodels para modelos VAR y VARMAX, pyflux para modelos bayesianos, y tensorflow para modelos basados en redes neuronales.
- En R: Paquetes como vars, tsDyn, y urca ofrecen funcionalidades avanzadas para análisis multivariado.
- En Power BI y Tableau: Pueden integrarse con R o Python para visualizar y modelar series de tiempo multivariadas de manera interactiva.




**Nuevas Tendencias y Técnicas en Series de Tiempo Multivariadas**

- Machine Learning: Algoritmos como Random Forest, XGBoost y redes neuronales recurrentes (RNNs) se aplican cada vez más en la predicción de series de tiempo multivariadas, ofreciendo capacidades de modelado más flexibles.
- Redes Neuronales Temporales (LSTM y GRU): Ideales para capturar dependencias a largo plazo en datos secuenciales.
- Modelos Bayesianos: Ofrecen un enfoque probabilístico que es útil para manejar la incertidumbre y la variabilidad inherentes en series multivariadas.

### Introducción al Modelo ARMAX (Autoregressive Moving Average with Exogenous Inputs)

El modelo ARMAX es una extensión del modelo ARMA (Autoregressive Moving Average) que incorpora variables exógenas (factores externos) para mejorar la capacidad predictiva y el ajuste del modelo. Es un enfoque muy útil cuando creemos que nuestra serie de tiempo está influenciada por factores externos que no se pueden capturar solo observando la serie de tiempo en sí.

####  **Comprendiendo los Componentes**

El modelo ARMAX combina tres elementos principales:

- **AR (Autoregresivo):** Usa los valores pasados de la serie para predecir el valor actual. Por ejemplo, el valor de la serie en el tiempo \(t\) puede depender de los valores en \(t-1\), \(t-2\), etc.
  
- **MA (Promedio Móvil):** Usa los errores pasados de la predicción para ajustar el valor actual. Esto significa que los errores cometidos en el pasado tienen un impacto en las predicciones futuras.

- **X (Variables Exógenas):** Incorpora variables externas que pueden influir en la serie temporal. Estas variables no son parte de la serie en sí, pero tienen un efecto sobre ella. Por ejemplo, si estamos modelando la demanda de electricidad, factores como la temperatura o el día de la semana podrían ser variables exógenas.

#### **La Ecuación del Modelo ARMAX**

La ecuación general del modelo ARMAX se expresa como:

$$
y_t = c + \sum_{i=1}^{p} \phi_i y_{t-i} + \sum_{j=1}^{q} \theta_j \epsilon_{t-j} + \sum_{k=1}^{r} \beta_k X_{t-k} + \epsilon_t
$$

Donde:
- $y_t$ es el valor de la serie en el tiempo $t$.
- $c$ es una constante.
- $p$ es el orden del componente autoregresivo (AR), y $\phi_i$ son los coeficientes asociados a los valores pasados $y_{t-i}$.
- $q$ es el orden del componente de promedio móvil (MA), y \(\theta_j\) son los coeficientes asociados a los errores pasados $\epsilon_{t-j}$.
- $r$ es el número de variables exógenas, y $\beta_k$ son los coeficientes de las variables exógenas $X_{t-k}$.
- $\epsilon_t$ es el término de error en el tiempo $t$.




#### **¿Por qué Utilizar un Modelo ARMAX?**

El modelo ARMAX es útil en situaciones donde las series temporales están influenciadas por factores externos. A diferencia del modelo ARMA, que solo se enfoca en la relación interna de la serie, el ARMAX puede capturar la influencia de factores adicionales que tienen un impacto directo en la serie temporal.



#### **Ejemplos de Aplicación del Modelo ARMAX**

- **Economía:** Para modelar la tasa de inflación utilizando el pasado de la inflación (AR), el impacto de choques económicos pasados (MA), y factores externos como la tasa de interés o el precio del petróleo (X).
- **Energía:** Predicción del consumo de energía utilizando la temperatura como una variable exógena.
- **Marketing:** Predicción de ventas utilizando variables como campañas publicitarias y estacionalidad como variables exógenas.




####  **Pasos para Estimar un Modelo ARMAX**

1. **Identificación:** Seleccionar los órdenes \(p\), \(q\), y \(r\) para el modelo.
   - Esto se puede hacer utilizando funciones de autocorrelación (ACF) y autocorrelación parcial (PACF), así como mediante la experimentación con diferentes combinaciones.
  
2. **Estimación:** Ajustar el modelo a los datos y estimar los parámetros \(\phi_i\), \(\theta_j\), y \(\beta_k\).

3. **Validación:** Verificar la validez del modelo comprobando los residuos para asegurarse de que no hay patrones no capturados.

4. **Pronóstico:** Utilizar el modelo para hacer predicciones sobre valores futuros de la serie.



### Series SOI y Recruitment

Las series de tiempo SOI (Southern Oscillation Index) y Recruitment son conocidas en el campo de la climatología y biología marina, y se utilizan para analizar la relación entre fenómenos climáticos y la biología de organismos marinos.

**SOI (Southern Oscillation Index)**
Descripción: El Índice de Oscilación del Sur (SOI) es una medida que representa las diferencias de presión atmosférica entre Tahití y Darwin (Australia). Es un indicador del fenómeno El Niño/Oscilación del Sur (ENSO), que tiene un impacto importante en el clima global.

Interpretación:
Valores positivos de SOI indican condiciones de La Niña (enfriamiento del Pacífico ecuatorial), que suelen asociarse con climas más fríos y húmedos en ciertas regiones.
Valores negativos de SOI indican condiciones de El Niño (calentamiento del Pacífico ecuatorial), que tienden a producir climas más cálidos y secos.



**Recruitment (Reclutamiento)**

Descripción: La serie "Recruitment" generalmente se refiere al número de nuevos individuos que se agregan a una población en un ecosistema marino, como la cantidad de peces juveniles que sobreviven y se unen a la población adulta.

Importancia: El reclutamiento es un indicador crucial para la gestión pesquera y la sostenibilidad de las especies marinas. Cambios en el reclutamiento a menudo reflejan variaciones en factores ambientales, como la temperatura del agua, el plancton, y eventos climáticos como El Niño y La Niña.
Relación entre SOI y Recruitment

El SOI afecta las condiciones oceánicas y, por ende, la disponibilidad de nutrientes y condiciones para el crecimiento de organismos marinos.
En particular, la influencia de El Niño y La Niña puede alterar la temperatura del océano, la disponibilidad de alimento y, por tanto, los niveles de reclutamiento de peces y otras especies.

Usos y Aplicación en Análisis de Series de Tiempo

Al analizar ambas series juntas, se puede estudiar cómo las condiciones climáticas impactan la dinámica de las poblaciones marinas, lo cual es vital para la planificación de la pesca, la conservación de especies y el entendimiento de los impactos del cambio climático en los ecosistemas marinos.

## Funciones de correlación cruzada y relaciones entre 2 series temporales


- El problema básico que estamos considerando es la descripción y el modelado de la relación entre dos series de tiempo.
  
 
- En la relación entre dos series de tiempo $y_t$ y $x_t$  La serie $y_t$ puede estar relacionado con retrasos pasados de la serie $x$ . 
  
  
- **La función de correlación cruzada de muestra (CCF)** es útil para identificar retrasos de la variable $x$ que podrían ser predictores útiles de $y_t$



En R, el CCF de muestra se define como el conjunto de correlaciones de muestra entre $x_{t+h}$ y $y_t$ para $h = 0, ± 1, ± 2, ± 3$, y así sucesivamente. Un valor negativo para $h$ es una correlación entre la $x-$variable  en un momento anterior y la variable $y$ en el tiempo $t$. 

*Por ejemplo, considere $h= −2$. El valor CCF daría la correlación entre $x_{t-2}$ y $y_t$ .


* Cuando uno o más $x_{t+h}$  , con $h$  negativo , son predictores de $y_t$, a veces se dice que $x$ lleva a $y$} .

* Cuando uno o más $x_{t+h}$ , con $h$ positivo , son predictores de $y_t$, a veces se dice que $x$ se retrasa $y$



En algunos problemas, el objetivo puede ser identificar qué variable está liderando y cuál está retrasada. 
    
* Sin embargo, en muchos problemas que consideramos, examinaremos las variables $x$ como una variable principal de la variable $y$ porque queremos utilizar los valores de la variable $x$ para predecir los valores futuros de $y$.

Por lo tanto, generalmente veremos qué sucede con los valores negativos de  en la parcela CCF.

```python
from statsmodels.tsa.stattools import ccf
import numpy as np
import matplotlib.pyplot as plt

# ccf(x, y, nlags=k): correlacion entre x[t] e y[t+k]
# lag k > 0 significa que SOI en el periodo t predice Recruitment k meses DESPUES.
#
# CORRECCION: la version anterior usaba np.correlate con una normalizacion
# incorrecta que inflaba las correlaciones en los rezagos extremos.
# statsmodels.ccf aplica el estimador muestral corregido (adjusted=True).

nlags      = 24
ccf_vals   = ccf(soi_rec['SOI'], soi_rec['Recruitment'], nlags=nlags, adjusted=True)
# lags_pos se ajusta dinamicamente al largo real devuelto por ccf
# (distintas versiones de statsmodels devuelven nlags o nlags+1 valores)
lags_pos   = np.arange(len(ccf_vals))
conf_bound = 1.96 / np.sqrt(len(soi_rec))   # banda de confianza aproximada al 95 %

fig, ax = plt.subplots(figsize=(14, 5))
colors = ['crimson' if abs(v) > conf_bound else 'steelblue' for v in ccf_vals]
ax.bar(lags_pos, ccf_vals, color=colors, width=0.55, alpha=0.85, edgecolor='white')
ax.axhline( conf_bound, color='gray', linestyle='--', linewidth=1.3,
            label=f'IC 95 % (+-{conf_bound:.3f})')
ax.axhline(-conf_bound, color='gray', linestyle='--', linewidth=1.3)
ax.axhline(0, color='black', linewidth=0.8)
ax.set_title(
    'Funcion de Correlacion Cruzada (CCF): SOI -> Recruitment
'
    'Barras en rojo: lags estadisticamente significativos  |CCF| > IC 95 %',
    fontsize=12)
ax.set_xlabel('Rezago k (meses) -- SOI precede a Recruitment en k periodos', fontsize=11)
ax.set_ylabel('Correlacion cruzada', fontsize=11)
ax.set_xticks(lags_pos)
ax.legend(fontsize=10)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Banda de confianza al 95 %: +- {conf_bound:.4f}
")
print(f"{'Lag':>5}  {'CCF':>9}  Significativo")
print("-" * 32)
for k, v in zip(lags_pos, ccf_vals):
    marca = " <-- SI" if abs(v) > conf_bound else ""
    print(f"{k:>5}  {v:>+9.4f}  {marca}")
```

### Interpretacion de la CCF

## Pruebas Formales de Estacionariedad

Antes de seleccionar los ordenes de diferenciacion del modelo SARIMA,
aplicamos las pruebas **ADF** y **KPSS** sobre el conjunto de **entrenamiento**.

| Prueba | H0 | Decision si p < 0.05 |
|--------|----|----------------------|
| **ADF** | La serie tiene raiz unitaria (NO estacionaria) | Rechazar H0 -> la serie **ES** estacionaria |
| **KPSS** | La serie **ES** estacionaria | Rechazar H0 -> la serie **NO** es estacionaria |

Usar las dos pruebas en conjunto reduce el riesgo de conclusiones erroneas.
Si **ambas** indican no-estacionariedad en nivel -> se aplica diferenciacion (d >= 1).

```python
from statsmodels.tsa.stattools import adfuller, kpss
import warnings
warnings.filterwarnings('ignore')

def test_estacionariedad(serie, nombre):
    """Ejecuta ADF y KPSS e imprime un resumen interpretable."""
    adf_stat, adf_p, _, _, _, _ = adfuller(serie.dropna(), autolag='AIC')
    kpss_stat, kpss_p, _, _    = kpss(serie.dropna(), regression='c', nlags='auto')
    dec_adf  = "Estacionaria       OK" if adf_p  < 0.05 else "NO estacionaria  (diferenciacion requerida)"
    dec_kpss = "Estacionaria       OK" if kpss_p > 0.05 else "NO estacionaria  (diferenciacion requerida)"
    print(f"{'─'*62}")
    print(f"  Serie : {nombre}")
    print(f"  ADF   : estadistico = {adf_stat:8.4f}   p = {adf_p:.4f}  ->  {dec_adf}")
    print(f"  KPSS  : estadistico = {kpss_stat:8.4f}   p = {kpss_p:.4f}  ->  {dec_kpss}")

print("NIVEL (serie original)\n")
test_estacionariedad(train_rec1,                "Recruitment")
test_estacionariedad(train_soi1,                "SOI")

print("\nPRIMERA DIFERENCIA REGULAR  (d = 1)\n")
test_estacionariedad(train_rec1.diff().dropna(), "Recruitment  [d=1]")
test_estacionariedad(train_soi1.diff().dropna(), "SOI  [d=1]")

print("\nCONCLUSION:")
print("  Si Recruitment no es estacionaria en nivel pero si con d=1 -> se justifica d=1.")
print("  SOI suele ser estacionario en nivel (sin necesidad de diferenciacion).")
```

### Ajustar el modelo ARIMA


**pm.auto_arima**
La función auto_arima intenta encontrar el mejor modelo ARIMA (Autorregresivo Integrado de Medias Móviles) ajustado a una serie temporal, seleccionando automáticamente los órdenes del modelo con base en criterios de información como AIC o BIC.

Argumentos

- train_rec1: Es la serie temporal de entrenamiento a la que se ajustará el modelo ARIMA.

- seasonal=True: Este argumento indica si se debe ajustar un componente estacional en el modelo (en este caso, es un modelo SARIMA, que es ARIMA con componente estacional).

- m=12: Es el número de períodos por ciclo estacional. Como m=12, el modelo considera una estacionalidad anual con datos mensuales (12 meses por año).

- d=1: Es el parámetro de diferenciación no estacional del modelo ARIMA. Es el número de veces que se debe diferenciar la serie temporal para hacerla estacionaria en términos no estacionales.

- D=1: Es el parámetro de diferenciación estacional. Es el número de veces que se debe diferenciar la serie para hacerla estacionaria en términos estacionales.

- start_p=1: Es el valor inicial para la búsqueda del parámetro p, que es el orden del componente autorregresivo (AR) no estacional.

- start_q=0: Es el valor inicial para la búsqueda del parámetro q, que es el orden del componente de medias móviles (MA) no estacional.

- max_p=2: Es el valor máximo permitido para p en el proceso de búsqueda automática del mejor modelo.

- max_q=2: Es el valor máximo permitido para q en el proceso de búsqueda automática del mejor modelo.

- start_P=0: Es el valor inicial para la búsqueda del parámetro P, que es el orden del componente autorregresivo (AR) estacional.

- start_Q=1: Es el valor inicial para la búsqueda del parámetro Q, que es el orden del componente de medias móviles (MA) estacional.

- max_P=2: Es el valor máximo permitido para P en la búsqueda automática.

- max_Q=2: Es el valor máximo permitido para Q en la búsqueda automática.

information_criterion="bic": El criterio de información a utilizar para seleccionar el mejor modelo. En este caso, se está utilizando el Criterio de Información Bayesiano (BIC), que penaliza modelos más complejos, favoreciendo aquellos más simples.

- trace=True: Muestra información de las iteraciones mientras se realiza la búsqueda de los mejores parámetros.

- error_action='ignore': Ignora ciertos errores comunes que pueden ocurrir durante el ajuste del modelo.

- suppress_warnings=True: Suprime las advertencias para evitar mensajes innecesarios durante el proceso.

- stepwise=True: Realiza la búsqueda de los mejores parámetros del modelo de manera iterativa (o paso a paso), lo que puede ser más rápido en comparación con una búsqueda exhaustiva de todas las combinaciones posibles.

### Diagnostico de Residuos — Modelo Univariado

Un modelo bien especificado debe producir residuos que se comporten como **ruido blanco**:
sin autocorrelacion, distribucion aproximadamente normal y varianza constante.

| Panel | Que evalua |
|-------|-----------|
| Residuos estandarizados | No deben mostrar tendencia ni heteroscedasticidad |
| Histograma + KDE | La distribucion debe ser aproximadamente gaussiana |
| QQ-plot | Los puntos deben seguir la diagonal |
| ACF de residuos | Ninguna barra debe superar las bandas de confianza |

La **prueba de Ljung-Box** (H0: los residuos son ruido blanco) confirma formalmente
si queda autocorrelacion significativa en los residuos.

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

# El objeto ARIMAResults de statsmodels esta disponible dentro de pmdarima via .arima_res_
fig = modelo.arima_res_.plot_diagnostics(figsize=(14, 10))
fig.suptitle('Diagnostico de Residuos -- SARIMA(1,1,0)(2,1,0)12  (Modelo Univariado)',
             fontsize=13, y=1.01)
plt.tight_layout()
plt.show()

# Prueba formal de Ljung-Box
lb_univ = acorr_ljungbox(modelo.arima_res_.resid, lags=[6, 12, 24], return_df=True)
lb_univ.index.name = 'Lag'
lb_univ = lb_univ.rename(columns={'lb_stat': 'Estadistico Q', 'lb_pvalue': 'p-valor'})
print("Prueba de Ljung-Box  (H0: residuos = ruido blanco)")
print(lb_univ.to_string())
print("\nInterpretacion: p > 0.05 -> no rechazamos H0 -> residuos sin autocorrelacion")
```

El pronóstico que un modelo puede hacer cuando utilizas variables exógenas rezagadas depende directamente del número de periodos que hayas especificado como rezagos. 

- Explicación del Pronóstico con Variables Exógenas Rezagadas
Cuando utilizas un modelo de series de tiempo con variables exógenas rezagadas, estás incluyendo información de la variable exógena proveniente de periodos anteriores para ayudar a predecir la variable dependiente en el presente.

Por ejemplo, al usar los rezagos de la variable exógena SOI con los periodos 5, 6, 7, 8 y 9, el modelo utilizará la información de la variable SOI de 5 a 9 periodos anteriores a cada punto de tiempo para hacer predicciones.

Limitaciones del Pronóstico
El modelo puede pronosticar hasta el número de periodos que sea igual al menor rezago que incluiste. En este caso, como el rezago mínimo es 5, el modelo podrá pronosticar hasta 5 periodos futuros sin requerir información adicional.

Si intentas pronosticar más allá de esos 5 periodos, el modelo ya no tendrá información disponible para los rezagos requeridos (5, 6, 7, 8 y 9), y por lo tanto no podrá hacer un pronóstico.

## Tabla Comparativa de Modelos

Comparamos los cuatro modelos estimados en este cuadernillo usando metricas
**fuera de muestra** (periodo de validacion: oct 1986 - sep 1987).

| Metrica | Formula | Interpretacion |
|---------|---------|----------------|
| **RMSE** | (1/n sum(yi - y_hat_i)^2)^0.5 | Error en unidades originales; penaliza errores grandes |
| **MAPE** | 100/n sum abs(yi-y_hat_i)/yi  | Error porcentual; facilita comparacion entre escalas |

> **Menor RMSE y MAPE -> mejor poder predictivo fuera de muestra.**

La inclusion de SOI como variable exogena debe mejorar las metricas respecto al modelo univariado.
La pregunta clave es cual especificacion del rezago resulta mas util para el pronostico.

```python
import pandas as pd
from sklearn.metrics import root_mean_squared_error, mean_absolute_percentage_error

modelos_pred = {
    'M0  SARIMA(1,1,0)(2,1,0)12  [univariado]' : forecast_series,
    'M1  SARIMAX + SOI(t)         [contemp.]'  : forecast_modelo1,
    'M2  SARIMAX + SOI(t-6)       [lag 6]'     : forecast_lag6,
    'M3  SARIMAX + SOI(t-5:t-9)   [multi-lag]' : forecast_multilag,
}

filas = []
for nombre, pred in modelos_pred.items():
    y_true = validation_rec1.reindex(pred.index).dropna()
    y_pred = pred.reindex(y_true.index)
    filas.append({
        'Modelo'  : nombre,
        'RMSE'    : root_mean_squared_error(y_true, y_pred),
        'MAPE (%)': mean_absolute_percentage_error(y_true, y_pred) * 100,
    })

tabla = pd.DataFrame(filas).set_index('Modelo').round(4)
tabla = tabla.sort_values('RMSE')

print("=" * 65)
print("    COMPARACION DE MODELOS  (menor RMSE/MAPE = mejor)")
print("=" * 65)
print(tabla.to_string())
print("=" * 65)
print(f"\nMejor modelo (menor RMSE): {tabla.index[0]}")
print("\nNota: metricas calculadas sobre el periodo de validacion (oct 1986 - sep 1987).")
```
